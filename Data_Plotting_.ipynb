# Use bar chart for plotting on missing values

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('titanic.csv')
data= df.isnull().sum()
data.plot(kind = "bar",color= "cyan")
plt.xlabel("attributes")
plt.ylabel("Missing values")
```

# Use pie plot for plotting the percentage of survival

```python

data = df['Survived'].value_counts()
print(data)
plt.pie(data, labels = data.index, autopct ="%1.1f%%", startangle=90, colors=['orange', 'cyan'])
plt.title("Survival Pie Chart")
```
