# Use bar chart for plotting on missing values

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
df = pd.read_csv('titanic.csv')
data= df.isnull().sum()
# data2=df.describe() 
# agemean=df['Age'].describe().loc['min']
# print(agemean)

data.plot(kind='bar',color='orangered')
#plt.xlabel("attributes")
#plt.ylabel("Missing values")
```

# Use pie plot for plotting the percentage of survival

```python

data = df['Survived'].value_counts()
print(data)
plt.pie(data, labels = data.index, autopct ="%1.3f%%", startangle=90, colors=['orange', 'cyan'])
plt.title("Survival Pie Chart")
```

```python
sb.histplot(data=df,hue='Survived',multiple='stack',x='Age',kde='True',bins=30)
```
