```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('./Book1.csv')

df
```

```python
plt.scatter(df['x'],df['y'])

plt.xlabel('x')
plt.ylabel('y')
```

```python
from sklearn.cluster import KMeans

kme=KMeans(n_clusters=2)

y_pred = kme.fit_predict(df[['x','y']])

print(y_pred)

print(kme.cluster_centers_)
```

```python
df['cluster']=y_pred

plt.scatter(df['x'],df['y'],c=df['cluster'],cmap='plasma',label="datapoints")
plt.scatter(kme.cluster_centers_[:,0],kme.cluster_centers_[:,1],marker='*',s=100,c='red',label='centeroide')

plt.title('kmeans on x y')

plt.xlabel('X')
plt.ylabel('Y')

plt.legend()
```
