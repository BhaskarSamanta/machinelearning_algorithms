```python
import pandas as pd
import numpy as np
import seaborn as sb
from sklearn.preprocessing import LabelEncoder
```

# 1) printing the data set in tabular format

```python
df=sb.load_dataset('titanic')
df 
```

<h1>Preeprocessing the data</h1>

```python
df['pclass'].unique()
```

```python
nan_columns = df.columns[df.isna().any()].tolist()
print("Columns with NaN values:", nan_columns)
```

```python
cols_to_drop=['parch','embarked','class','who','adult_male','deck','embark_town','alive','alone']
df.drop(columns=cols_to_drop,inplace=True)
df 
```

<h1>Normalizing the Data Set</h1>

```python
neumer_colls=['pclass','age','sibsp','fare']
df[neumer_colls]=(df[neumer_colls]-df[neumer_colls].min())/(df[neumer_colls].max()-df[neumer_colls].min())

df
```

<h1>Encoding nonnuemaric values to numaric values</h1>

```python
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df
```

<h1>deleating the touples containing nan values</h1>

```python
rowsBefore=df.shape[0]
df.dropna(inplace=True)
rowsAfter=df.shape[0]
print("before: ",rowsBefore,"\nafter: ",rowsAfter,"\ntotal nan tuples: ",rowsBefore-rowsAfter)
```

<h1>Preeprocessing has finished... spliting the dataset</h1>

```python
from sklearn.model_selection import train_test_split
x=df[['pclass','sex','age','sibsp','fare']]
y=df['survived']

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=1)
print('value of xtraint: \n',x_train.shape,'\nvalue of xtest: \n',x_test.shape,'\nvalue of ytraint: \n',y_train.shape,'\nvalue of ytest: \n',y_test.shape)
```

# applying knn

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report 
from sklearn.metrics import accuracy_score

knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train,y_train)
pred=knn.predict(x_test)

rpt=classification_report(pred,y_test)

asc=accuracy_score(pred,y_test)

print('detail report: \n',rpt,'\naccuracy: \n',asc)
```
