```python
import pandas as pd 
import seaborn as sns

df = sns.load_dataset('titanic')
df
```

```python
cols_to_drop =['embarked','who','adult_male','deck','embark_town','alive','alone','class']

df = df.drop(columns= cols_to_drop,axis = 1)
df
```

```python
df = df.dropna()
df
```

```python
from sklearn.preprocessing import LabelEncoder

le =LabelEncoder()

df['sex'] = le.fit_transform(df['sex'])

df
```

```python
from sklearn.model_selection import train_test_split

x = df[['pclass','sex','age','sibsp','parch','fare']]
y = df['survived']

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state= 2)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report

model = LogisticRegression()

model.fit(x_train,y_train)

predd = model.predict(x_test)
# aivhave intersecpt kore
inter = model.intercept_
print(inter)

accu = accuracy_score(predd,y_test)
rpt = classification_report(predd,y_test)
print('The Accuracy is: \n',accu,'\n' ,'Detailed Report',rpt)
```
