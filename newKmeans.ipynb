```python
import pandas as pd 
import matplotlib.pyplot as plt
data1={'X':[1,1,2,3,2,3,2,11,11,12,13,12,13,12],
       'Y':[1,2,1,2,3,3,2,1,2,1,2,3,3,2]}

df=pd.DataFrame(data1)
df
```

```python
df.to_csv('newDF.csv',index=False)

df=dict(df)
print(df)
df=pd.DataFrame(df)

df
```

```python
data=df

data
```

```python
plt.scatter(data['X'],data['Y'])

plt.xlabel('X')
plt.ylabel('Y')


```

```python
from sklearn.cluster import KMeans
# from sklearn.metrics import accuracy_score 
 
km=KMeans(n_clusters=2)
pred=km.fit_predict(data[['X','Y']])

print(pred)
print(km.cluster_centers_)
```

```python
plt.scatter(data['X'],data['Y'],c=pred,label='datapoints')#,cmap='rainbow')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker='*',label='centeroid',s=100,c='red')

plt.legend()
```
